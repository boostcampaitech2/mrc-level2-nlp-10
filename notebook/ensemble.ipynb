```python
import json
```

```python
with open('../outputs/test_dataset/nbest_predictions_1.json') as f:
    nbest_1 = json.load(f)

with open('../outputs/test_dataset/nbest_predictions_2.json') as f:
    nbest_2 = json.load(f)

with open('../outputs/test_dataset/nbest_predictions_3.json') as f:
    nbest_3 = json.load(f)
```

## nbest_1

```python
query_id = list(nbest_1.keys())
```

```python
import heapq

dic = {}
for id in query_id:
    pq = []
    temp = nbest_1.get(id)
    for v in temp: pq.append((-v['probability'], v['text']))
    temp = nbest_2.get(id)
    for v in temp: pq.append((-v['probability'], v['text']))
    temp = nbest_3.get(id)
    for v in temp: pq.append((-v['probability'], v['text']))
    heapq.heapify(pq)
    _, text = heapq.heappop(pq)
    dic[id]=text
```

```python
dic
```

```python
with open('../outputs/test_dataset/predictions.json', 'w', encoding="utf-8") as writer:
    writer.write(
        json.dumps(dic, indent=4, ensure_ascii=False) + '\n'
        )
```
