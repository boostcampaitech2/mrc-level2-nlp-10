```python
import json
```

```python
with open('../outputs/test_dataset/nbest_predictions_roberta.json') as f:
    roberta = json.load(f)

with open('../outputs/test_dataset/nbest_predictions_bert.json') as f:
    bert = json.load(f)

with open('../outputs/test_dataset/nbest_predictions_electra.json') as f:
    electra = json.load(f)
```

## nbest_1

```python
query_id = list(roberta.keys())
```

```python
import heapq
dic = {}
for id in query_id:
    roberta_ = roberta.get(id)
    bert_ = bert.get(id)
    electra_ = electra.get(id)

    roberta_first = roberta_[0]['text']; roberta_prob = roberta_[0]['probability']
    bert_first = bert_[0]['text']; bert_prob = bert_[0]['probability']
    electra_first = electra_[0]['text']; electra_prob = electra_[0]['probability']

    pq = [(-roberta_prob,roberta_first),(-bert_prob,bert_first),(-electra_prob,electra_first)]
    heapq.heapify(pq)
    _, text = heapq.heappop(pq)
    
    m = {}

    if roberta_first not in m: m[roberta_first] = 1
    else: m[roberta_first] += 1 * 2
    
    if bert_first not in m: m[bert_first] = 1
    else: m[bert_first] += 1

    if electra_first not in m: m[electra_first] = 1
    else: m[electra_first] += 1 * 1.5

    if len(m) == 3:
        dic[id]=text
    else:
        sorted(m.items())
        dic[id]=list(m)[0]


    
    
```

```python
with open('../outputs/test_dataset/predictions.json', 'w', encoding="utf-8") as writer:
    writer.write(
        json.dumps(dic, indent=4, ensure_ascii=False) + '\n'
        )
```
